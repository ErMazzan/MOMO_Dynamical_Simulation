```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
filename = "dt_10-4/initial_distribution.dat"
values = read_file(filename,0)
f, ax = plt.subplots()
plt.hist(values, range=[-20,20], density = True, bins=30)

f.set_size_inches(4, 6.5)
plt.rc('xtick', labelsize=12)
plt.rc('ytick', labelsize=12)
plt.xlabel("V", fontsize=15)
plt.ylabel("Density", fontsize=15)
plt.show()
```

```python
filename = "dt_10-4/initial_distribution.dat"
values = read_file(filename,0)
f, (ax1, ax2) = plt.subplots(1,2)
ax1.hist(values, range=[-20,20], density = True, bins=30)

plt.rc('xtick', labelsize=14) 
plt.rc('ytick', labelsize=14)

ax1.set_xlabel("Velocity (a.u.)", fontsize=14)
ax1.set_ylabel(r"$\rho$ (a.u.)", fontsize=14)

filename = "dt_10-4/final_distribution.dat"
values = read_file(filename,0)
ax2.hist(values, range=[-30,30], density = True, bins=30, label="Velocities")

def g(x):
    T = 100
    return 1/(np.sqrt(1.5*np.pi*T))*np.exp(-x*x/(2*T))
x=np.linspace(-30,30,101)

ax2.plot(x, g(x),color="red", linewidth=3.0, label="Fitting")
f.set_size_inches(6, 6.5)

f.suptitle('dt = $10^{-4}$', x=0.55, y=0.98, size=20.0)

ax2.legend(loc="upper left", prop={'size': 12})
ax2.set_xlabel("Velocity (a.u.)", fontsize=14)
ax2.set_ylabel(r"$\rho$ (a.u.)", fontsize=14)
ax2.set_ylim([0.0, 0.06])

fig = plt.gcf()
fig.tight_layout(pad=.0)
fig.set_size_inches(10, 6)
plt.subplots_adjust(top=0.88)
plt.savefig('velocities_distribution-dt-10-4.pdf')  
plt.show()
```

```python
def read_file(filename, column):
    with open(filename, 'r') as data:
        x = []
        for line in data:
            p = line.split()
            x.append(float(p[column]))

    return x
```
